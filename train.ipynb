```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_df = pd.read_csv('../data/FlightDelay/train.csv')
test_df = pd.read_csv('../data/FlightDelay/test.csv')
```

```python
train_df
```

```python
train_df.isnull().sum().to_frame()
```

```python
train_df = train_df.dropna()
train = TabularDataset(train_df.drop(['ID'], axis=1))
test = TabularDataset(test_df.drop(['ID'], axis=1))
```

```python
predictor = TabularPredictor(label='Delay', eval_metric='log_loss',).fit(train)
```

```python
predictor.leaderboard(train, silent=True)
```

```python
predictor.feature_importance(train)
```

```python
pred_y = predictor.predict_proba(test)
```

```python
submission = pd.read_csv('../data/FlightDelay/sample_submission.csv')
submission['Not_Delayed'], submission['Delayed'] = pred_y['Not_Delayed'], pred_y['Delayed']

submission.to_csv('submission01.csv', index=False)
```
